# OpenCV (HAAR)

Import the required libraries

```python
from matplotlib import pyplot as plt
import cv2

# plot inline
%matplotlib inline

# larger plots
plt.rcParams['figure.figsize'] = (16,8)

# https://matplotlib.org/gallery/style_sheets/style_sheets_reference.html
plt.style.use('seaborn-whitegrid')
```

## Demo Face detection

Show an original example

```python
def show_image(img, title):
    plt.title(title)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.xticks([])
    plt.yticks([]) 
    plt.show()

# load an image
img = cv2.imread('../data/raw/DahQw7YXcAo67yn.jpg')

# show the image
show_image(img, 'Original Image')
```

Detect all faces

```python
# clone the original image
clone = img.copy()

# Get the face detector
face_cascade = cv2.CascadeClassifier('../models/haarcascade_frontalface_default.xml')

# find all faces in the image
for (x,y,w,h) in face_cascade.detectMultiScale(cv2.cvtColor(clone, cv2.COLOR_BGR2GRAY), 1.3, 5):
    cv2.rectangle(clone, (x,y), (x + w, y + h), (255,0,0), 2)

# show the image
show_image(clone, 'Faces found')
```
